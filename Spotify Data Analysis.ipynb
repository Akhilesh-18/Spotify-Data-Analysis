```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_tracks=pd.read_csv(r"C:\Users\akhil\Downloads\spotifydatasets\tracks.csv")
data_tracks.head()
```

```python
pd.isnull(data_tracks).sum()
```

```python
data_tracks.info()
```

```python
sorted_tracks=data_tracks.sort_values('popularity',ascending=True).head(10)
sorted_tracks
```

```python
data_tracks.describe().transpose()
```

```python
most_popular=data_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
data_tracks.set_index('release_date', inplace = True)
data_tracks.index=pd.to_datetime(data_tracks.index)
data_tracks.head()
```

```python
data_tracks[["artists"]].iloc[25]
```

```python
data_tracks['duration'] = data_tracks['duration_ms'].apply (lambda x : round(x/1000))
data_tracks.drop('duration_ms', inplace = True, axis=1)
data_tracks.duration.head()
```

```python
corr_tracks = data_tracks.drop(['key','mode','explicit'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
map = sns.heatmap(corr_tracks, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Correlation Map')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
sample_tracks = data_tracks.sample(int(0.004 * len(data_tracks)))
len(sample_tracks)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_tracks, y='loudness', x='energy', color='r').set(title='Loudness Vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_tracks, y='popularity', x='acousticness', color='y').set(title='Popularity Vs Acousticness Correlation')
```

```python
data_tracks['dates']=data_tracks.index.get_level_values('release_date')
data_tracks.dates=pd.to_datetime(data_tracks.dates)
years=data_tracks.dates.dt.year
```

```python
data_tracks.head()
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
total_dr = data_tracks.duration
fig_dims = (18,6)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```
